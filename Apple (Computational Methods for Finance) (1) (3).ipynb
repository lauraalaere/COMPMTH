Data Input of the European put option

```python
S0 = data[-1:]  # Spot Price = 192.02
K = 192.02      # Strike Price
T = 1           # Maturity 1 year
r = 0.05        # Risk Free Rate 
sig = std_dev   # Diffucion coefficient / Volatility 
N = 3           # Number of periods or number of time periods
payoff = "put"  # Payoff

print(S0)
```

# The Binomial Tree Model

```python
dT = float(T) / N                             # Delta t
u = np.exp(sig * np.sqrt(dT))                 # up factor
d = 1.0 / u                                   # down factor

S = np.zeros((N + 1, N + 1))
S[0, 0] = float(S0.iloc[0])
z = 1
for t in range(1, N + 1):        #looping forwards, from 1 to N
    for i in range(z):           #looping forwards, from 0 to z-1
        S[i, t] = S[i, t-1] * u
        S[i+1, t] = S[i, t-1] * d
    z += 1  

print('The binomial tree presenting Apple price over 3 time steps','\n',(S))
```

```python
print('The up factor u is',(u))
```

```python
print('The down factor d is',(d))
```

```python
a = np.exp(r * dT)    # risk free compound return
p = (a - d)/ (u - d)  # risk neutral up probability
q = 1.0 - p           # risk neutral down probability
print ('The risk-neutral probability is',(p))
```

# Option value at each final nodes 

```python
S_T = S[:,-1]
V = np.zeros((N + 1, N + 1))
if payoff =="call":
    V[:,-1] = np.maximum(S_T-K, 0.0)
elif payoff =="put":
    V[:,-1] = np.maximum(K-S_T, 0.0)
V
```

# Option value at earlier nodes

```python
# for European Option
for j in range(N-1, -1, -1): # Column. looping backwards. From N-1 to 0
   for i in range(j+1):  # Row. looping forwards. From 0 to j
       V[i,j] = np.exp(-r*dT) * (p * V[i,j + 1] + q * V[i + 1,j + 1]) #the theoretical value at each node.
V

print('European ' + payoff, str( V[0,0]))
```

```python
# for American Option
if payoff =="put":
    for j in range(N-1, -1, -1):
       for i in range(j+1):
           V[i,j] = np.maximum(S[i,j] - K,np.exp(-r*dT) * (p * V[i,j + 1] + q * V[i + 1,j + 1])) 
           #compare the payoffs from early exercise with the theoretical option value. 
elif payoff =="call":
    for j in range(N-1, -1, -1):
       for i in range(j+1):
           V[i,j] = np.maximum(K - S[i,j],np.exp(-r*dT) * (p * V[i,j + 1] + q * V[i + 1,j + 1]))
V
```

```python
print ('American' + payoff, str(V[0,0]))
```

# Monte Carlo Simulation 

```python
def mcs_simulation_np(p):
    M = p
    I = p
    dt = T / M 
    S = np.zeros((M + 1, I))
    S[0] = S0 
    rn = np.random.standard_normal(S.shape) 
    for t in range(1, M + 1): 
        S[t] = S[t-1] * np.exp((r - sig ** 2 / 2) * dt + sig * np.sqrt(dt) * rn[t])    
    return S
```

```python
S2 = mcs_simulation_np(1000)
```

# Delta

```python
S0 = 192.02     # Spot Price
K = 192.02      # Strike Price
T = 1           # Maturity 1 year
r = 0.05        # Risk Free Rate 
sig = 0.2908   # Diffusion coefficient / Volatility 
N = 3           # Number of periods or number of time periods
payoff = "put"  # Payoff
```

```python
def delta(S, K, T, r, q, sig, payoff):
    d1 = (np.log(S / K) + (r - q + 0.5 * sig ** 2) * T) / (sig * np.sqrt(T))
    if payoff == "call":
        delta = np.exp(-q * T) * norm.cdf(d1, 0.0, 1.0)
    elif payoff == "put":
        delta = -np.exp(-q * T) * norm.cdf(-d1, 0.0, 1.0)
    return delta
```

```python
delta(S0, K, T, r , 0, sig, 'put')
```

```python
S11 = np.linspace(150, 250, 50)
Delta_Call = np.zeros((len(S11),1))
Delta_Put = np.zeros((len(S11),1))
for i in range(len(S11)):
    Delta_Put [i] = delta(S11[i], K, T, r, 0, sig, 'put')
```

```python
fig = plt.figure()

plt.plot(S11, Delta_Put, '-')
plt.grid()
plt.xlabel('Apple Price')
plt.ylabel('Delta')
plt.title('Delta')
plt.legend(['Delta for Put'])
```

# Gamma 

```python
S0 = 192.02     # Spot Price 
K = 192.02      # Strike Price
T = 1           # Maturity 1 year
r = 0.05        # Risk Free Rate 
sig = 0.2908   # Diffusion coefficient / Volatility 
N = 3           # Number of periods or number of time periods
q = 0
payoff = "put"  # Payoff
```

```python
def gamma(S, K, T, r, vol, payoff):
    d1 = (np.log(S / K) + (r - q + 0.5 * vol ** 2) * T) / (vol * np.sqrt(T))
    gamma = np.exp(-q * T) * norm.pdf(d1, 0.0, 1.0) / (S * vol * np.sqrt(T))
    
    
    return gamma
```

# Theta

```python
def theta(S, K, T, r, vol, payoff):
    d1 = (np.log(S / K) + (r - q + 0.5 * sig ** 2) * T) / (vol * np.sqrt(T))
    d2 = d1 - vol * np.sqrt(T)
    if payoff == "call":
        theta = -S * np.exp(-q * T) * norm.pdf(d1, 0.0, 1.0) * vol / (2 * np.sqrt(T)) - q * K * np.exp(-r * T) * norm.cdf(d2, 0.0, 1.0)
    elif payoff == "put":
        theta = -S * np.exp(-q * T) * norm.pdf(d1, 0.0, 1.0) * vol / (2 * np.sqrt(T)) + q * K * np.exp(-r * T) * norm.cdf(-d2, 0.0, 1.0)
    return theta
```

```python
theta(192.02, 192.02, 1, 0.05 , 0.2908, 'put')
```

```python
fig = plt.figure()
plt.plot(T, Theta_Put, '-')
plt.grid()
plt.xlabel('Time to Expiry')
plt.ylabel('Theta')
plt.title('Theta')
plt.legend(['Theta for Put'])
```

# Vega

```python
def vega(S, K, T, r, std_dev, payoff):
    
    d1 = (np.log(S / K) + (r + 0.5 * std_dev ** 2) * T) / (std_dev * np.sqrt(T))
    N_d1_prime=1/np.sqrt(2 * np.pi) * np.exp(-d1**2/2)
    vega = S * np.sqrt(T) * N_d1_prime
    
    return vega
```

```python
vega(192.02, 192.02, 1, 0.05 , 0.2908, 'put')
```

```python
vol = np.linspace(0.1,0.4,13)
Vega = np.zeros((len(vol),1))
for i in range(len(vol)):
    Vega [i] = vega(192.02, 192.02, 1, 0.05, vol[i], 'put')
```
